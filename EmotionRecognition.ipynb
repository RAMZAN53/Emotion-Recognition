```python
from tensorflow.keras.models import load_model
import numpy as np
import os
import librosa
import pickle
```

```python
from google.colab import drive
drive.mount('/content/drive',force_remount=True)
```

```python
def LoadModelAndEncoder():
    model = load_model('/content/drive/MyDrive/TASKS/EmotionRecognitionModel.h5')
    with open('/content/drive/MyDrive/TASKS/EmotionRecognitionModel.pkl', 'rb') as f:
        le = pickle.load(f)
    return model, le
```

```python
def LoadAndExtract(file_path):
    y, sr = librosa.load(file_path, sr=None)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
    mfccs = np.mean(mfccs.T, axis=0)
    return mfccs
```

```python
def Prediction(file_path, model, le):
    features = LoadAndExtract(file_path)
    features = features[np.newaxis, ..., np.newaxis]
    prediction = model.predict(features)
    predicted_label = le.inverse_transform([np.argmax(prediction)])
    return predicted_label[0]
```

```python
model, le = LoadModelAndEncoder()
file = '/content/drive/MyDrive/TASKS/Voice1.wav'
Predicted_Emotion = Prediction(file, model, le)
print(f'Predicted Emotion :: {Predicted_Emotion}')
```
